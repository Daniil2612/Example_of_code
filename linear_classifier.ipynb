```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


df=pd.read_csv("/content/sample_data/weatherAUS_train.csv")
df_ = pd.read_csv("/content/sample_data/weatherAUS_test.csv")

df_['RainToday'] = df_['RainToday'].map({'Yes': 1, 'No': 0})
df['RainToday'] = df['RainToday'].map({'Yes': 1, 'No': 0})

y = df["RainTomorrow"]
df.head()
```

```python
df_.shape
```

```python
pd.set_option('display.max_columns', None)
df.columns
```

```python

numeric_data = list(df.dtypes[df.dtypes == "float64"].index)
categorical = list(df.dtypes[df.dtypes == "object"].index)
df = pd.concat([df[numeric_data],
                     pd.get_dummies(df.Location),
                     pd.get_dummies(df.WindGustDir),
                     pd.get_dummies(df.WindDir9am),
                     pd.get_dummies(df.WindDir3pm)], axis=1)
df.head()
```

```python
df["RainTomorrow"] = y
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split

numeric_data = df.select_dtypes([np.number])
numeric_data = numeric_data.drop('RainTomorrow', axis=1)
Y=df["RainTomorrow"]
X=df.drop(columns=["RainTomorrow"])

##X.insert(1, "Bias", np.ones((len(X), 1)), True)

X_train , X_test, Y_train, Y_test =  train_test_split(X, Y, test_size=0.4, random_state=42)


numeric_data_med = numeric_data.median()
numeric_features = numeric_data.columns

X_train = X_train.fillna(numeric_data_med)

X_test = X_test.fillna(numeric_data_med)

#categorical = list(X_train.dtypes[X_train.dtypes == "object"].index)

## X = np.hstack([X, np.ones((len(X), 1))])
X_train.head()
```

Normalization

```python
scaler = StandardScaler(copy = False)
X_train_scaled = scaler.fit(X_train[numeric_features])
X_test_scaled = scaler.transform(X_test[numeric_features])
```

We made normal data, then we can work with it

```python

#instantiate the model
logreg = LogisticRegression(solver='liblinear', random_state=0)


# fit the model
logreg.fit(X_train, Y_train)
#prediction
y_pred_test = logreg.predict(X_test)

y_pred_test
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(Y_test, y_pred_test)))
```
